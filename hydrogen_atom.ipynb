```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def probability_1s(x, y, z):
  r = np.sqrt(np.square(x) + np.square(y) + np.square(z))
  return np.square(np.exp(-r) / np.sqrt(np.pi))
```

```python
x = np.linspace(0,1,30)
y = np.linspace(0,1,30)
z = np.linspace(0,1,30)
```

```python
elements = []
probability = []
```

```python
for i in x:
  for j in y:
    for k in x:
      elements.append(str((i, j, k)))
      probability.append(probability_1s(i, j, k))

  
probability_sum = probability / sum(probability)
probability_sum
```

```python
coordinates = np.random.choice(elements, size=100000, replace=True, p=probability_sum)
element_matrix = [i.split(',') for i in coordinates]
element_matrix = np.matrix(element_matrix)
x_coordinates = [float(i.item()[1:]) for i in element_matrix[:,0]] 
y_coordinates = [float(i.item()) for i in element_matrix[:,1]]
z_coordinates = [float(i.item()[0:-1]) for i in element_matrix[:,2]]
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_coordinates, y_coordinates, z_coordinates, alpha=0.1, s=1)
ax.set_title("Hydrogen")
plt.show()
```
